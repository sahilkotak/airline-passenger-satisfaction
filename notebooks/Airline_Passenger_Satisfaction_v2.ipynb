### Data Preparation

- Importing the test set.

```python
test_set = test_set = pd.read_csv('/Users/varunsinghchhetri/Downloads/archive (7) 2/test.csv')
```

```python
test_set.columns=[each.replace(" ","_") for each in test_set.columns]
test_set.columns=[each.replace("/","_") for each in test_set.columns]
test_set.columns=[each.replace("-","_") for each in test_set.columns]
```

- Filling the missing values in 'Arrival_Delay_in_Minutes' by mode.

```python
test_set["Arrival_Delay_in_Minutes"]=test_set["Arrival_Delay_in_Minutes"].fillna(stats.mode(test_set["Arrival_Delay_in_Minutes"]).mode[0])

```

- Dropping the columns 'id' and 'Unnamed:_0'

```python
test_set = test_set.drop(['id','Unnamed:_0'],axis=1)
```

- Mapping the values in categorical columns to numerical values for modelling purposes

```python
def data_preparation(df):

    mappings = {
        'Gender': {'Female': 1, 'Male': 0},
        'Customer_Type': {'Loyal Customer': 1, 'disloyal Customer': 0},
        'Type_of_Travel': {'Business travel': 1, 'Personal Travel': 0},
        'Class': {'Business': 2, 'Eco Plus': 1, 'Eco': 0},
        'satisfaction': {'satisfied': 1, 'neutral or dissatisfied': 0}
    }
    
    for col, mapping in mappings.items():
        df[col] = df[col].map(mapping).fillna(-1)
    
    return df
```

```python
df_train = data_preparation(data)
df_test = data_preparation(test_set)
```

### Splitting the Data into X_train, y_train, X_test and y_test

```python
# Separate features and target variable from train_set
X_train = df_train.drop('satisfaction', axis=1)
y_train = df_train['satisfaction'].to_numpy()
```

```python
# Separate features from test_set
X_test = df_test.drop('satisfaction', axis=1)
y_test = df_test['satisfaction'].to_numpy()
```

### Scaling the X data using StandardScaler

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

### Modelling 

- For the given Data, we will be using 5 different Algorithms
- 1. Logistic Regressor Classifier
- 2. Random Forest Classifier
- 3. Gradient Boost 
- 4. AdaBoost Classifier 
- 5. XGBoost Classifier

- The evaluation metric will be based on AUROC Score and Confusion Matrix
- Why AUROC:


- 1. It measures the ability of the model to distinguish between the positive and negative classes. A higher AUROC score indicates better distinguishing capability.
- 2. It evaluates the model's performance across all possible thresholds, giving us a better picture, rather than to a specific one. This is very useful when you are not sure what threshold would be most appropriate for classification.
- 3. It allows us to compare the performance of different models on the same problem, helping us to understand which model best fits the dataset.




- Why Confusion Matrix:


- 1. The AUROC gives us a single metric which summarizes the model's ability to distinguish between classes, the confusion matrix gives us a more clear view of the model's performance.
- 2. The confusion matrix helps us understand the type of errors that our model is making. Eg; it shows us if our model is more prone to false positives or false negatives, which could be crucial to gain insights about the nature of our dataset.

#### 1. Logistic Regressor Classifier

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, roc_curve, confusion_matrix, auc
```
